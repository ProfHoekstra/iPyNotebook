# Sympy basics: Sympy as your math assistant

Here's a simple program to solve $ax^2+bx+c=0$ :

```python
print "Oplossen ax2 + bx + c = 0 :"
#a,b,c =  input("Geef a,b,c:")
a,b,c = 1,7,-2

D = b*b-4*a*c
if D<0:
    print "Er zijn geen oplossingen voor x"
    
elif D==0:
    print "Er is 1 oplossing: x = ",-b/(2.*a)
    
else:
    print "Oplossing 1: x = ", (-b-sqrt(D))/(2.*a)
    print "Oplossing 2: x = ", (-b+sqrt(D))/(2.*a)
```

This is the type of solution you would also find with a calculator. Using Sympy you can use 'symbolic math', just like the program Mathematica or the Wolfram Alpha website (which driven by Mathematica):

```python

from sympy import *
init_printing()
x=symbols('x')

print "Oplossen ax2 + bx + c = 0 :"
#a,b,c =  input("Geef a,b,c:")
a,b,c = 1,7,-2
 
D = b*b-4*a*c
if D<0:
    print "Er zijn geen oplossingen voor x"
    
elif D==0:
    print "Er is 1 oplossing: x = ",-b/(2.*a)
    
else:
    print "Oplossing 1: x = ", (-b-sqrt(D))/(2.*a)
    print "Oplossing 2: x = ", (-b+sqrt(D))/(2.*a)
   
```

Using symbols, much more is possible, like differentiation, integration and solving

```python
solve(x*x+7*x-2)
```

```python
x,y=symbols('x y')
diff(exp(x)/sin(x))
```

```python
x/sqrt(x*x+y*y)
```

```python
integrate(x/sqrt(x*x+y*y),x)
```

```python
solve(x*x*+7*x*x-2*x+5)
```

```python
s=solve(x*x*x+7*x*x-2*x+5)
```

```python
complex(s[0]) # float() would not work here as it's a complex number
```

Solving integrals

Compute the integral: $ \int_{-\infty}^\infty {sin(x^2)} {dx}$
Let's first see how this function looks.

(Note how we now suddenly need to specify explicitly to use numpy and matplotlib as sympy has been imported as star, sowe are in the namespace of Sympy.)

```python
import matplotlib.pyplot as plt
import numpy
xtab = linspace(-10,10,100)
ytab = numpy.sin(xtab*xtab)
plt.plot(xtab,ytab)
```

To compute the integral: $ \int_{-\infty}^\infty {sin(x^2)} {dx}$ add a tuple with (symbol, startvalue, stopvalue)

```python
integrate(sin(x**2), (x, -oo, oo))
```

```python
float(integrate(sin(x**2), (x, -oo, oo)))
```
